# Split the dataset into a training set and a validation set

```python
src_folder = 'detection_dataset_v3'
dst_folder = 'detection_dataset_v4'

if not Path(dst_folder).is_dir(): os.mkdir(dst_folder)
```

```python
sliceslist = os.listdir(src_folder)

rmslices = []
sliceNumList = []
for cur_slice in sliceslist:
    tmp_path = os.path.join(dst_folder, cur_slice)
    filelist = os.listdir(os.path.join(src_folder, cur_slice))
    tmp_num = len(filelist)
    sliceNumList += [tmp_num]
    if tmp_num>3:
        rmslices += [cur_slice]
    else:
        if not Path(tmp_path).is_dir(): os.mkdir(tmp_path)
        for cur_file in filelist:
            src_path = os.path.join(src_folder, cur_slice, cur_file)
            dst_path = os.path.join(dst_folder, cur_slice, cur_file)
            shutil.copyfile(src_path, dst_path)
```
